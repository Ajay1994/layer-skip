```python
import json
import os
def get_json_list(file_path):
    file_path = os.path.expanduser(file_path)
    with open(file_path, "r") as f:
        json_list = []
        for line in f:
            if len(line) > 100:
                json_list.append(json.loads(line))
        return json_list
```
