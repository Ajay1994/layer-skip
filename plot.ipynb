```python
fig, axs = plt.subplots(1, 4,  figsize = (25, 5))
l_width = 3.0
x_label = 18
y_label = 18
legend_size = 14
title_size = 18
marker_size = 50
last = 15

data = pd.read_csv("commit_logs/cosine_sim_wikitext_7B.txt", sep = ",", header = None).iloc[:, :32]
for i in range(0, len(data)):
    row = [float(x) for x in list(data.iloc[i, :])]
    axs[0].plot(np.arange(len(row)), row, alpha=0.1, color="wheat")
row = [float(x) for x in list(data.iloc[0, :])]
axs[0].plot(np.arange(len(row)), row, alpha=0.9, color="seagreen", linestyle="--", label = "First Token")
row = [float(x) for x in list(data.iloc[-1, :])]
axs[0].plot(np.arange(len(row)), row, alpha=0.9, color="darkorange",linestyle="--", label = "Last Token")
data = data.astype(float)
mean = list(data.mean(axis=0))
axs[0].plot(np.arange(len(row)), mean, color="crimson", label = "Mean")
axs[0].axvspan(0, 10, alpha=0.2, color='lightcoral')
axs[0].axvspan(29, 32, alpha=0.2, color='lightcoral')
axs[0].axvspan(10, 29, alpha=0.1, color='mediumspringgreen')
axs[0].tick_params(axis='both',  labelsize=14)
axs[0].set_xlabel("Layer Number", fontsize=x_label)
axs[0].set_ylabel("Cosine Similarity", fontsize=y_label)
axs[0].legend(fontsize= legend_size)
axs[0].set_title("LLama2-7B | C4", fontsize=title_size)
axs[0].set_ylim([0.6, 1])


data = pd.read_csv("commit_logs/cosine_sim_c4_7B.txt", sep = ",", header = None).iloc[:, :32]
for i in range(0, len(data)):
    row = [float(x) for x in list(data.iloc[i, :])]
    axs[1].plot(np.arange(len(row)), row, alpha=0.1, color="wheat")
row = [float(x) for x in list(data.iloc[0, :])]
axs[1].plot(np.arange(len(row)), row, alpha=0.9, color="seagreen", linestyle="--", label = "First Token")
row = [float(x) for x in list(data.iloc[-1, :])]
axs[1].plot(np.arange(len(row)), row, alpha=0.9, color="darkorange",linestyle="--", label = "Last Token")
data = data.astype(float)
mean = list(data.mean(axis=0))
axs[1].plot(np.arange(len(row)), mean, color="crimson", label = "Mean")
axs[1].axvspan(0, 10, alpha=0.2, color='lightcoral')
axs[1].axvspan(29, 32, alpha=0.2, color='lightcoral')
axs[1].axvspan(10, 29, alpha=0.1, color='mediumspringgreen')
axs[1].tick_params(axis='both',  labelsize=14)
axs[1].set_xlabel("Layer Number", fontsize=x_label)
axs[1].set_ylabel("Cosine Similarity", fontsize=y_label)
axs[1].legend(fontsize= legend_size)
axs[1].set_title("LLama2-7B | C4", fontsize=title_size)
axs[1].set_ylim([0.6, 1])



data = pd.read_csv("commit_logs/cosine_sim_wikitext_13B.txt", sep = ",", header = None).iloc[:, :40]
for i in range(0, len(data)):
    row = [float(x) for x in list(data.iloc[i, :])]
    axs[2].plot(np.arange(len(row)), row, alpha=0.2, color="wheat")
row = [float(x) for x in list(data.iloc[0, :])]
axs[2].plot(np.arange(len(row)), row, alpha=0.9, color="seagreen", linestyle="--", label = "First Token")
row = [float(x) for x in list(data.iloc[-1, :])]
axs[2].plot(np.arange(len(row)), row, alpha=0.9, color="darkorange",linestyle="--", label = "Last Token")
data = data.astype(float)
mean = list(data.mean(axis=0))
axs[2].plot(np.arange(len(row)), mean, color="crimson", label = "Mean")
axs[2].axvspan(0, 10, alpha=0.2, color='lightcoral')
axs[2].axvspan(35, 40, alpha=0.2, color='lightcoral')
axs[2].axvspan(10, 35, alpha=0.1, color='mediumspringgreen')
axs[2].tick_params(axis='both',  labelsize=14)
axs[2].set_xlabel("Layer Number", fontsize=x_label)
axs[2].set_ylabel("Cosine Similarity", fontsize=y_label)
axs[2].legend(fontsize= legend_size)
axs[2].set_title("LLama2-13B | WikiText", fontsize=title_size)
axs[2].set_ylim([0.6, 1])



data = pd.read_csv("commit_logs/cosine_sim_c4_13B.txt", sep = ",", header = None).iloc[:, :40]
for i in range(0, len(data)):
    row = [float(x) for x in list(data.iloc[i, :])]
    axs[3].plot(np.arange(len(row)), row, alpha=0.2, color="wheat")
row = [float(x) for x in list(data.iloc[0, :])]
axs[3].plot(np.arange(len(row)), row, alpha=0.9, color="seagreen", linestyle="--", label = "First Token")
row = [float(x) for x in list(data.iloc[-1, :])]
axs[3].plot(np.arange(len(row)), row, alpha=0.9, color="darkorange",linestyle="--", label = "Last Token")
data = data.astype(float)
mean = list(data.mean(axis=0))
axs[3].plot(np.arange(len(row)), mean, color="crimson", label = "Mean")
axs[3].axvspan(0, 10, alpha=0.2, color='lightcoral')
axs[3].axvspan(35, 40, alpha=0.2, color='lightcoral')
axs[3].axvspan(10, 35, alpha=0.1, color='mediumspringgreen')
axs[3].tick_params(axis='both',  labelsize=14)
axs[3].set_xlabel("Layer Number", fontsize=x_label)
axs[3].set_ylabel("Cosine Similarity", fontsize=y_label)
axs[3].legend(fontsize= legend_size)
axs[3].set_title("LLama2-13B | C4", fontsize=title_size)
axs[3].set_ylim([0.6, 1])

plt.tight_layout()
plt.savefig("plots/fnn_skip_analysis.pdf")
```
