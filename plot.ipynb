```python
import json
import os
def get_json_list(file_path):
    file_path = os.path.expanduser(file_path)
    with open(file_path, "r") as f:
        json_list = []
        for line in f:
            if len(line) > 100:
                json_list.append(json.loads(line))
        return json_list
```

```python
questions = get_json_list("table/question.jsonl")
categories = [questions[i]["category"] for i in range(0, len(questions))]
response_1012141210 = get_json_list("commit_logs/judge_result000000_1012141210.jsonl")
```

```python
score_unpruned = []
score_pruned = [] 
for item in response_1012141210:
    score = item["score"]
    score_unpruned.append(score[0])
    score_pruned.append(score[1])
```

```python
result = pd.DataFrame()
result["category"] = categories
result["pruned"] = score_pruned
result["unpruned"] = score_unpruned
unique_cat = np.unique(categories)
```

```python
mask = result['pruned'] == -1.0
result = result[~mask]
```
